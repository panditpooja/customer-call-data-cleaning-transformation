```python
# Step 1: Define mapping of common timezone abbreviations to full pytz timezone names
tz_map = {
    'IST': 'Asia/Kolkata',   # India Standard Time
    'MST': 'US/Mountain',    # Mountain Standard Time
    'PST': 'US/Pacific',     # Pacific Standard Time
    'EST': 'US/Eastern',     # Eastern Standard Time
    'CET': 'Europe/Berlin'   # Central European Time
}

# Step 2: Remove ordinal suffixes like 2nd → 2
def remove_ordinal_suffix(date_val):
    return re.sub(r'(\d{1,2})(st|nd|rd|th)', r'\1', str(date_val)) #re.sub() replaces only the matched substring, not the full string.


# Function to standardize and parse a date string or timestamp into UTC format
def standardize_and_parse(date_val):
    try:
        # Clean up the date string (remove ordinal suffixes like '2nd', strip whitespace)
        cleaned = remove_ordinal_suffix(str(date_val).strip())

        # Split the string into parts to check for any timezone abbreviation at the end
        parts = cleaned.split()
        tz_abbr = parts[-1] if parts[-1] in tz_map else None  # Look for abbreviation
        tz_name = tz_map.get(tz_abbr)  # Get corresponding full timezone name

        # If a known timezone abbreviation is present
        if tz_name:
            # Remove the abbreviation from the string to safely parse the datetime
            cleaned = cleaned.replace(tz_abbr, '').strip()

            # Parse the cleaned date string into a naive datetime (no timezone yet)
            dt_naive = parser.parse(cleaned, dayfirst=False)

            # Localize to the detected timezone and convert to UTC
            dt_local = pytz.timezone(tz_name).localize(dt_naive)
            dt = dt_local.astimezone(pytz.utc)

        # If no timezone info is found, assume it's UTC or parse naively
        else:
            # Try parsing assuming MM/DD/YYYY format
            try:
                dt = parser.parse(cleaned, dayfirst=False)
            # Fallback: if that fails, try DD/MM/YYYY format
            except:
                dt = parser.parse(cleaned, dayfirst=True)

            # Set timezone to UTC manually since it's a naive datetime
            dt = dt.replace(tzinfo=pytz.utc)

        # Return the final datetime as a string in standardized UTC format
        return dt.strftime('%Y-%m-%d %H:%M:%S')

    # If any unexpected error occurs, print it and return None
    except Exception as e:
        print(f"[Failed] {date_val} → {e}")
        return None

# Apply the function to the 'Date' column in the DataFrame to create a new 'Standardized_UTC' column
df['Standardized_UTC_Date'] = df['Date'].apply(standardize_and_parse)

# View the original and parsed columns
df[['Date', 'Standardized_UTC_Date']]
```
